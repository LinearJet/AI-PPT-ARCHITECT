Let's generate your first image with the loaded Stable Diffusion XL model! We'll use a simple prompt and then display the result.

```python
import os
import torch
from diffusers import StableDiffusionXLPipeline

image_models_path = '/content/drive/My Drive/Image models'
safetensor_path = os.path.join(image_models_path, 'waiIllustrious.safetensors')

print(f"Attempting to load model from: {safetensor_path}")
print(f"Does the path exist? {os.path.exists(safetensor_path)}")
print(f"Is it a file? {os.path.isfile(safetensor_path)}")

pipeline = StableDiffusionXLPipeline.from_single_file(
    safetensor_path,
    torch_dtype=torch.float16,
    use_safetensors=True
).to('cuda')
```

Now, let's set up some parameters for generating images. You can adjust these to experiment with different styles and qualities:

*   **`prompt`**: The text prompt describing the image you want to generate.
*   **`num_inference_steps`**: The number of denoising steps. Higher values generally lead to better quality but take longer.
*   **`guidance_scale` (CFG Scale)**: Controls how strongly the image generation process adheres to the prompt. Higher values make the AI follow the prompt more strictly, but can sometimes lead to less creative results.
*   **`scheduler`**: The algorithm used to denoise the image. Different schedulers can produce distinct visual characteristics and vary in speed.

```python

from diffusers import (
    EulerDiscreteScheduler,
    EulerAncestralDiscreteScheduler,
    DPMSolverMultistepScheduler,
    UniPCMultistepScheduler
)

# UI sliders stay exactly the same:
num_inference_steps = 32  # @param {type:"slider", min: 10, max: 100, step: 1}
guidance_scale = 7      # @param {type:"slider", min: 0.0, max: 20.0, step: 0.5}

# NEW CLEAN DROPDOWN:
selected_scheduler = "Euler a"  # @param ["Euler", "Euler a", "DPM++ 2M", "UniPC"]

# NEW KARRAS TOGGLE:
use_karras = True  # @param {type:"boolean"}

SCHEDULERS = {
    "Euler": EulerDiscreteScheduler,
    "Euler a": EulerAncestralDiscreteScheduler,
    "DPM++ 2M": DPMSolverMultistepScheduler,
    "UniPC": UniPCMultistepScheduler,
}

# Apply scheduler cleanly:
scheduler_class = SCHEDULERS[selected_scheduler]
pipeline.scheduler = scheduler_class.from_config(
    pipeline.scheduler.config,
    **({"use_karras_sigmas": True} if use_karras else {})
)

print(f"Sampler: {selected_scheduler} {'(Karras)' if use_karras else ''}")
print(f"Steps: {num_inference_steps}")
print(f"CFG: {guidance_scale}")
```

## Download LoRA from Civitai

This section allows you to download a LoRA model directly from Civitai by providing its URL or ID. The downloaded `.safetensors` file will be saved to your `Loras` folder in Google Drive.

```python
import requests
from tqdm.auto import tqdm
import re
import os

def download_civitai_lora(model_url_or_id, output_dir):
    """
    Downloads a LoRA file from Civitai using its URL or ID.

    Args:
        model_url_or_id (str): The Civitai URL or ID of the LoRA model.
        output_dir (str): The directory where the LoRA file will be saved.
    """
    print(f"Attempting to download LoRA from: {model_url_or_id}")

    model_id = None
    if isinstance(model_url_or_id, int):
        model_id = model_url_or_id
    elif isinstance(model_url_or_id, str):
        match = re.search(r'civitai\.com/models/(\d+)', model_url_or_id)
        if match:
            model_id = int(match.group(1))
        elif model_url_or_id.isdigit():
            model_id = int(model_url_or_id)

    if not model_id:
        print(f"Error: Invalid Civitai URL or ID provided: {model_url_or_id}")
        return

    # Civitai API endpoint for model details
    api_url = f"https://civitai.com/api/v1/models/{model_id}"

    try:
        response = requests.get(api_url, timeout=10)
        response.raise_for_status() # Raise an exception for HTTP errors
        model_data = response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching model details from Civitai API: {e}")
        return

    # Find the primary file and its download URL and filename
    download_url = None
    file_name = None

    # Civitai API structure can vary, typically 'modelVersions' contains file info
    if 'modelVersions' in model_data and model_data['modelVersions']:
        # Assuming the first model version's primary file is what we want
        # You might need more sophisticated logic for specific versions/files
        first_version = model_data['modelVersions'][0]
        if 'files' in first_version and first_version['files']:
            for file_info in first_version['files']:
                if file_info.get('type') == 'Model' and file_info.get('primary'): # Check for primary model file
                    download_url = file_info.get('downloadUrl')
                    file_name = file_info.get('name')
                    break
            # If no primary file is found, just take the first model file
            if not download_url and not file_name:
                for file_info in first_version['files']:
                    if file_info.get('type') == 'Model':
                        download_url = file_info.get('downloadUrl')
                        file_name = file_info.get('name')
                        break

    if not download_url or not file_name:
        print(f"Error: Could not find download URL or filename for model ID {model_id}.")
        print("Please check if the model has a downloadable file or specify a different version if applicable.")
        return

    output_filepath = os.path.join(output_dir, file_name)

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created directory: {output_dir}")

    print(f"Downloading '{file_name}' to '{output_filepath}'...")

    try:
        with requests.get(download_url, stream=True, timeout=30) as r:
            r.raise_for_status() # Raise an exception for HTTP errors
            total_size_in_bytes = int(r.headers.get('content-length', 0))
            block_size = 1024 # 1 Kibibyte
            progress_bar = tqdm(total=total_size_in_bytes, unit='iB', unit_scale=True)
            with open(output_filepath, 'wb') as f:
                for chunk in r.iter_content(chunk_size=block_size):
                    progress_bar.update(len(chunk))
                    f.write(chunk)
            progress_bar.close()
            print(f"Successfully downloaded '{file_name}' to '{output_filepath}'")
    except requests.exceptions.RequestException as e:
        print(f"Error during file download: {e}")
        # Clean up partial file if download failed
        if os.path.exists(output_filepath):
            os.remove(output_filepath)
        return
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        if os.path.exists(output_filepath):
            os.remove(output_filepath)
        return

    return file_name
```

## Load Downloaded LoRA

Once a LoRA is downloaded, you can load it into your `pipeline` to apply its style or characteristics to image generation.

```python

import os
import warnings

# Define loras_path if not already defined
if 'loras_path' not in locals() or loras_path is None:
    base_drive_path = '/content/drive/My Drive'
    loras_path = os.path.join(base_drive_path, 'Loras')
    if not os.path.exists(loras_path):
        os.makedirs(loras_path)

# Unload any previously loaded LoRAs
if hasattr(pipeline, 'unload_lora_weights'):
    pipeline.unload_lora_weights()
    print("Unloaded any previously active LoRAs.")

# Load ALL LoRAs from the folder with error handling
lora_files = [f for f in os.listdir(loras_path) if f.endswith('.safetensors')]

print(f"Found {len(lora_files)} LoRA(s) in {loras_path}")

successfully_loaded = []
failed_to_load = []

# Suppress warnings during LoRA loading
warnings.filterwarnings('ignore')

for lora_file in lora_files:
    lora_base_name = os.path.splitext(lora_file)[0]
    full_lora_path = os.path.join(loras_path, lora_file)

    try:
        print(f"Loading LoRA: {lora_base_name}...", end=" ")

        # Redirect stderr to suppress the verbose error messages
        import sys
        from io import StringIO
        old_stderr = sys.stderr
        sys.stderr = StringIO()

        pipeline.load_lora_weights(
            full_lora_path,
            adapter_name=lora_base_name
        )

        # Restore stderr
        sys.stderr = old_stderr

        successfully_loaded.append(lora_base_name)
        print("✓")
    except Exception as e:
        # Restore stderr in case of error
        sys.stderr = old_stderr

        failed_to_load.append(lora_base_name)
        print(f"✗")

# Restore warnings
warnings.filterwarnings('default')

print(f"\n{'='*50}")
print(f"✓ Successfully loaded {len(successfully_loaded)} LoRA(s):")
for name in successfully_loaded:
    print(f"  • {name}")

if failed_to_load:
    print(f"\n✗ Skipped {len(failed_to_load)} incompatible LoRA(s):")
    for name in failed_to_load:
        print(f"  • {name}")
    print("\n  (These LoRAs are for a different model architecture)")

print(f"{'='*50}")
print("\nUse <lora:name:weight> in prompts to activate LoRAs.")
```

## Generate Image with Selected LoRA

This section allows you to define your prompt, select one of your downloaded LoRAs to apply, and set its strength. The selected LoRA will be applied for the next image generation.

```python
import re

# Text prompt input box - Using BOTH LoRAs!
prompt = ""  # @param {type:"string"}

# Parse LoRA tags from prompt
lora_pattern = r'<lora:([^:]+):([0-9.]+)>'
lora_matches = re.findall(lora_pattern, prompt)

# Remove LoRA tags from the actual prompt
clean_prompt = re.sub(lora_pattern, '', prompt).strip()
# Clean up any double spaces
clean_prompt = re.sub(r'\s+', ' ', clean_prompt)

# Set adapter weights based on tags
adapter_weights = {}
if lora_matches:
    print("LoRAs detected in prompt:")
    for lora_name, weight in lora_matches:
        adapter_weights[lora_name] = float(weight)
        print(f"  - {lora_name}: {weight}")

    # Set the adapter weights
    pipeline.set_adapters(list(adapter_weights.keys()), adapter_weights=list(adapter_weights.values()))
else:
    print("No LoRA tags found in prompt. Generating with base model only.")
    # Disable all adapters
    pipeline.disable_lora()

print(f"\nCleaned prompt: {clean_prompt}")

# Generate the image
image = pipeline(
    prompt=clean_prompt,
    num_inference_steps=num_inference_steps,
    guidance_scale=guidance_scale
).images[0]

# Display the image
display(image)
```
